```python
#this cell just imports PYQG and numpy
# https://pyqg.readthedocs.io/en/latest/examples/two-layer.html

import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import pyqg
```

```python
year = 24*60*60*360.
m = pyqg.QGModel(delta = 0.1, nx=128, tmax=20*year, twrite=5000, tavestart=5*year)
m2= pyqg.QGModel(delta = 1.5, nx=128, tmax=20*year, twrite=5000, tavestart=5*year)
m.run()
#m2.run()

plt.figure(figsize=(18,4)) 
#this runs the two-layer model with a delta(layer thickness ratio) of 0.1
q_upper = m.q[0] + m.Qy[0]*m.y 
plt.subplot(121)
plt.contourf(m.x, m.y, q_upper, 12, cmap='RdBu_r')
plt.xlabel('x'); plt.ylabel('y'); plt.title('Upper Layer PV with delta = 0.1')
plt.colorbar();



#this runs the two-layer model with a delta(layer thickness ratio) of 1 
year = 24*60*60*360.
m = pyqg.QGModel(delta = 1, nx=128, tmax=20*year, twrite=5000, tavestart=5*year)
m.run()

q_upper = m.q[0] + m.Qy[0]*m.y #this line 
plt.subplot(122)
plt.contourf(m.x, m.y, q_upper, 12, cmap='RdBu_r')
plt.xlabel('x'); plt.ylabel('y'); plt.title('Upper Layer PV with delta = 1')
plt.colorbar();

```
